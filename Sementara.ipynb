```python
#jangan select all waktu copas ke sini. 


import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("10besarpenyakitrawatjalan.csv")

#cara mengetahui ukuran file
#sumber (https://www.it-swarm.net/id/python/bagaimana-cara-memeriksa-ukuran-file-dengan-python/968266140/)
import os

def convert_bytes(num):
    """
    this function will convert bytes to MB.... GB... etc
    """
    for x in ['bytes', 'KB', 'MB', 'GB', 'TB']:
        if num < 1024.0:
            return "%3.1f %s" % (num, x)
        num /= 1024.0


def file_size(file_path):
    file_info = os.stat(file_path)
    return convert_bytes(file_info.st_size)

file_path = r"10besarpenyakitrawatjalan.csv"
hasil = file_size(file_path)
print("ukuran file =",hasil)

#cara mengetahui atribut data, panjang kolom dan baris set data
print(data.columns)
print("kolom =",len(data.columns))
print("baris =",len(data))

#Sampel Data
#10 baris pertama
data[0:10]

#Statistik Data

#Rata-rata
#Data dari jumlah kunjungan
data_kunjungan = data["Jumlah Kunjungan"]

#Data dari jumlah kasus baru
data_kasus = data["Jumlah Kasus Baru(4+5)"]

#Data dari jumlah kasus baru laki-laki
data_kasus_lk = data["Kasus Baru menurut Jenis Kelamin LK"]

#Data dari jumlah kasus baru perempuan
data_kasus_pr = data["Kasus Baru menurut Jenis Kelamin PR"]

#Rata-rata
#Data Kunjungan
print("Rata-rata jumlah kunjungan:")
print(data_kunjungan.mean())

#Data Jumlah Kasus Baru
print("Rata-rata jumlah kasus baru:")
print(data_kasus.mean())

#Data Jumlah Kasus Baru Laki-laki
print("Rata-rata jumlah kasus baru laki-laki:")
print(data_kasus_lk.mean())

#Data Jumlah Kasus Baru Perempuan
print("Rata-rata jumlah kasus baru perempuan:")
print(data_kasus_pr.mean())

#Standar Deviasi
#Standar deviasi dari jumlah kunjungan
print("Standar deviasi dari jumlah kunjungan:")
print(data["Jumlah Kunjungan"].std())

#Standar deviasi dari jumlah kasus baru
print("Standar deviasi dari jumlah kasus baru:")
print(data["Jumlah Kasus Baru(4+5)"].std())

#Standar deviasi dari jumlah kasus baru laki-laki
print("Standar deviasi dari jumlah kasus baru laki-laki:")
print(data["Kasus Baru menurut Jenis Kelamin LK"].std())

#Standar deviasi dari jumlah kasus baru perempuan
print("Standar deviasi dari jumlah kasus baru perempuan:")
print(data["Kasus Baru menurut Jenis Kelamin PR"].std())

#Percentile
#10%
#Percentile 10% dari jumlah kunjungan
print("Percentile 10% dari jumlah kunjungan:")
print(data_kunjungan.quantile(0.1))

#Percentile 10& dari jumlah kasus baru
print("Percentile 10% dari jumlah kasus baru:")
print(data_kasus.quantile(0.1))

#Percentile 10% dari kasus baru laki-laki
print("Percentile 10% dari jumlah kasus baru laki-laki:")
print(data_kasus_lk.quantile(0.1))

#Percentile 10% dari kasus baru permpuan
print("Percentile 10% dari jumlah kasus baru perempuan:")
print(data_kasus_pr.quantile(0.1))

#25%
#Percentile 25% dari jumlah kunjungan
print("Percentile 25% dari jumlah kunjungan:")
print(data_kunjungan.quantile(0.25))

#Percentile 25% dari jumlah kasus baru
print("Percentile 25% dari jumlah kasus baru:")
print(data_kasus.quantile(0.25))

#Percentile 25% dari kasus baru laki-laki
print("Percentile 25% dari jumlah kasus baru laki-laki:")
print(data_kasus_lk.quantile(0.25))

#Percentile 25% dari kasus baru perempuan
print("Percentile 25% dari jumlah kasus baru perempuan:")
print(data_kasus_pr.quantile(0.25))

#50%
#Percentile 50% dari jumlah kunjungan
print("Percentile 50% dari jumlah kunjungan:")
print(data_kunjungan.quantile(0.5))

#Percentile 50% dari jumlah kasus baru
print("Percentile 50% dari jumlah kunjungan:")
print(data_kunjungan.quantile(0.5))

#Ekstrimum
#Nilai maksismum dari jumlah kunjungan
imax = data["Jumlah Kunjungan"].idxmax()
data[imax:imax+1]

#Nilai maksimum dari jumlah kasus baru
imax = data["Jumlah Kasus Baru(4+5)"].idxmax()
data[imax:imax+1]

#Nilai maksismum dari kasus baru laki-laki
imax = data["Kasus Baru menurut Jenis Kelamin LK"].idxmax()
data[imax:imax+1]

#Nilai maksimum dari kasus baru perempuan
imax = data["Kasus Baru menurut Jenis Kelamin PR"].idxmax()
data[imax:imax+1]

#Nilai minimum dari jumlah kunjungan
imin = data["Jumlah Kunjungan"].idxmin()
data[imin:imin+1]

#Nilai minimum dari jumlah kasus baru
imin = data["Jumlah Kasus Baru(4+5)"].idxmin()
data[imin:imin+1]

#Nilai minimum dari kasus baru laki-laki
imin = data["Kasus Baru menurut Jenis Kelamin LK"].idxmin()
data[imin:imin+1]

#Nilai minimum dari kasus baru perempuan
imin = data["Kasus Baru menurut Jenis Kelamin PR"].idxmin()
data[imin:imin+1]


#Frekuensi
```

```python

#contoh perbandingan distribusi frekuensi
#masalahnya data bulan tersorting

stack = data.groupby([data["Bulan"],data["Deskripsi"].loc[(data["Deskripsi"]=="Hyperbilirubin")|(data["Deskripsi"]=="Olygohidramnion")]])["Jumlah Kunjungan"].sum().unstack()
print(stack)
stack.plot.line(title="Perbandingan data kunjungan Hyperbilirubin dan Olygohidramnion setiap bulannya",stacked=True)


```
